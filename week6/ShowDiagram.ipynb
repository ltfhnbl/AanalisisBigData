```python
from matplotlib import pyplot as plt
from matplotlib import style

x = [0, 1, 2, 3, 4, 5]
y = [24.27, 23.18, 22.19, 8.41, 7.19, 5.62]

fig, ax = plt.subplots()

ax.bar(x, y, align='center')

ax.set_title('Bahasa Pemograman Favorit (UMSU)')
ax.set_ylabel('Pengguna')
ax.set_xlabel('Bahasa')

ax.set_xticks(x)
ax.set_xticklabels(("Python", "JavaScript", "Java", "Ca", "PHP", "C++"))

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Teknisi', 'Guru', 'Pegawai Kantor', 'Pedagang']
quantity = [3, 10, 12, 11]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

plt.title("Daftar Pekerjaan Lulusan Darmajaya 2022")
plt.pie(quantity, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=30)

plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/sample_data/kemiskinan.csv')

provinsi = data['Provinsi']
kemiskinan = data['Kemiskinan']

fig, ax = plt.subplots(figsize=(10, 8))

ax.bar(provinsi, kemiskinan, align='center')

ax.set_title('Data Kemiskinan Diagram Batang')
ax.set_ylabel('Kemiskinan')
ax.set_xlabel('Provinsi')

ax.set_xticks(provinsi)
ax.set_xticklabels(provinsi, rotation=90)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/sample_data/kemiskinan.csv')

provinsi = data['Provinsi']
kemiskinan = data['Kemiskinan']

indeks_terpotong =  7

besaran_terpotong = 0.3

explode = [besaran_terpotong if i == indeks_terpotong else 0 for i in range(len(provinsi))]

fig, ax = plt.subplots(figsize=(10, 8))

ax.pie(kemiskinan, labels=provinsi, autopct='%1.1f%%', startangle=90, explode=explode)

ax.set_title('Data Kemiskinan Indonesia')

plt.axis('equal')

plt.show()
```
